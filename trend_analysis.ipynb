For missing value in Ratings, we will impute median rating value of each category.

```python
df['Category'].unique()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}, inplace=True)
```

```python
df.columns
```

Create a feature 'Age_Gropu' to analyse different age groups

```python
lables = ['Young Adults','Aduts','Middle-aged','Senior']
```

```python
df['age_group'] = pd.qcut(df['age'],q=4,labels=lables)
```

```python
df.head()
```

Create a feature 'Purchase Frequency Days' to analyse purchase patterns and frequency

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_maping = {'Fortnightly':14,
                    'Weekly':7,
                    'Annually':365,
                    'Quarterly':90,
                    'Monthly':30,
                    'Bi-Weekly':14,
                    'Every 3 Months':90
}
```

```python
df['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_maping)
```

```python
df[['frequency_of_purchases','Purchase_frequency_days']].head(10)
```

As discount applied column and promo code used colun as similar, let's check if they are same or not

```python
(df['promo_code_used']==df['discount_applied']).all()
```

Hence, delete one column

```python
df.drop(columns=['promo_code_used'], inplace=True)
```
